# Falsa posici√≥n

```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
plt.style.use(['science', 'notebook'])
```

```python
# definimos la funci√≥n
fx = lambda x: x**2-5*x+1
# graficamos para ver la forma de la curva
x = np.linspace(-2,7,100)
F = fx(x)
plt.plot(x,F, linestyle='-', color='magenta', label='f(x)')
plt.grid()
plt.axis([-2, 7, -6, 15])
plt.title('Curva de la funci√≥n')

# Graficamos la linea en y = 0 para visualizar los puntos de corte
x1 = np.linspace(-2,7,100)
cero = 0*x1
plt.plot(x1,cero, linestyle='-', color='blue', label='f(x)')
```

```python
# tolerancia o error
tolerancia = 10**(-10)

# Procedimiento

def falsa_posici√≥n(fx, a, b, tolerancia):
    tramo = abs(b-a)
    while tramo > tolerancia:
        fa = fx(a)
        fb = fx(b)
        c = b-fb*(b-a)/(fb-fa) # Valor de c
        fc = fx(c)
        
        # Verificamos si el valor de fc*fa es positivo o negativo
        cambio = np.sign(fa)*np.sign(fc)
        if cambio > 0:
            tramo = abs(c-a)
            a = c
        else :
            tramo = abs(b-c)
            b = c

    return print('La soluci√≥n es: ', c)
```

```python
falsa_posici√≥n(fx, 4, 5, tolerancia)
```

```python
falsa_posici√≥n(fx, 0, 1, tolerancia)
```
